```python
import yfinance as yf
import os
import pandas as pd
```

```python
btc_ticker = yf.Ticker("BTC-INR")
```

```python
if os.path.exists("btc.csv"):
    btc = pd.read_csv("btc.csv", index_col=0)
else:
    btc = btc_ticker.history(period="max")
    btc.to_csv("btc.csv")
```

```python
btc.index = (
    pd.to_datetime(btc.index)
      .tz_convert(None)     # remove timezone
      .normalize()          # set time to 00:00:00
)
```

```python
btc
```

```python
del btc["Dividends"]
del btc["Stock Splits"]
```

```python
btc.columns = [c.lower() for c in btc.columns]
```

```python
btc.plot.line(y="close", use_index=True)
```

```python
wiki = pd.read_csv("wikipedia_edits.csv", index_col=0, parse_dates=True)
```

```python
wiki
```

```python
btc = btc.merge(wiki, left_index=True, right_index=True)
```

```python
btc
```

```python
btc["tomorrow"] = btc["close"].shift(-1)
```

```python
btc
```

```python
btc["target"] = (btc["tomorrow"] > btc["close"]).astype(int)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=50, random_state=1)

train = btc.iloc[:-200]
test = btc.iloc[-200:]

predictors = ["close", "volume", "open", "high", "low", "edit_count", "sentiment", "neg_sentiment"]
model.fit(train[predictors], train["target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
preds = pd.Series(preds, index=test.index)
precision_score(test["target"], preds)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="predictions")
    combined = pd.concat([test["target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=1095, step=150):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    
    return pd.concat(all_predictions)
```

```python
def compute_rolling(btc):
    horizons = [2, 7, 60, 365]

    # Drop neg_sentiment — weak signal
    new_predictors = ["close", "sentiment"]

    for horizon in horizons:
        rolling_mean = (
            btc["close"]
            .rolling(horizon, min_periods=horizon)
            .mean()
            .shift(1)
        )

        ratio_column = f"close_ratio_{horizon}"
        btc[ratio_column] = btc["close"] / rolling_mean

        edit_column = f"edit_{horizon}"
        btc[edit_column] = (
            btc["edit_count"]
            .rolling(horizon, min_periods=horizon)
            .mean()
            .shift(1)
        )

        trend_column = f"trend_{horizon}"
        btc[trend_column] = (
            btc["target"]
            .rolling(horizon, closed="left", min_periods=horizon)
            .mean()
        )

        vol_column = f"volatility_{horizon}"
        btc[vol_column] = (
            btc["close"]
            .pct_change()
            .rolling(horizon, min_periods=horizon)
            .std()
            .shift(1)
        )

        new_predictors += [
            ratio_column,
            edit_column,
            trend_column,
            vol_column
        ]

    return btc, new_predictors
```

```python
btc, new_predictors = compute_rolling(btc.copy())
```

```python
predictions = backtest(btc, model, new_predictors)
```

```python
precision_score(predictions["target"], predictions["predictions"])
```
